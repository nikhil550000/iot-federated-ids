```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as pyplot
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
import os
```

```python
data_path = "/content/drive/MyDrive/iot_federated_ids/data/raw/Smart_Cam_Combined.csv"
df = pd.read_csv(data_path)
```

```python
df = df[['Power (Watts)','Label']]
```

```python
# Normalize the feature column
scaler = MinMaxScaler()
df["Power (Watts)"] = scaler.fit_transform(df[["Power (Watts)"]])
```

```python
values = df.values
np.random.shuffle(values)
df = pd.DataFrame(values, columns=df.columns)
```

```python
# Split into train (80%) and test (20%)
train_df, test_df = train_test_split(df, test_size=0.2, random_state=42)
```

```python
# Save processed data
processed_dir = "/content/drive/MyDrive/iot_federated_ids/data/processed/"
test_dir = "/content/drive/MyDrive/iot_federated_ids/data/test/"
os.makedirs(processed_dir, exist_ok=True)
os.makedirs(test_dir, exist_ok=True)
train_df.to_csv(f"{processed_dir}processed.csv", index=False)
test_df.to_csv(f"{test_dir}test.csv", index=False)
```

```python
# Split train data into clients (simulate 10 IoT devices)
client_dir = "/content/drive/MyDrive/iot_federated_ids/data/clients/"
os.makedirs(client_dir, exist_ok=True)
num_clients = 10
client_dfs = np.array_split(train_df, num_clients)
for i, client_df in enumerate(client_dfs):
    client_df.to_csv(f"{client_dir}client_{i+1}.csv", index=False)

print("Data preprocessing complete!")
```
